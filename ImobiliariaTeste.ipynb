```python
class Imovel:
    def __init__(self,data_contrato,dias_locados,inquilino,telefone):

        self.data_contrato = data_contrato
        self.dias_locados = dias_locados
        self.inquilino = inquilino
        self.telefone = telefone

    def retorna_data(self):
        return str(self.data_contrato)

    def recebe_dias_locados(self):
        return str(self.dias_locados)

    def recebe_inquilino(self):
        return str(self.inquilino)

    def recebe_telefone(self):
        return str(self.telefone)
```

```python
from datetime import datetime
import json

class Node:

    def __init__(self,imovel):
        self.right_node = None
        self.left_node = None
        self.imovel = imovel
        self.height = 1

    def print_tree_info(self, level=0):
        if self.right_node:
            self.right_node.print_tree_info(level + 1)

        print("  " * level + f"Data: {self.imovel.retorna_data()}, Inquilino: {self.imovel.recebe_inquilino()}, Altura: {self.height}, Balanceamento: {self.balance_factor()}")

        if self.left_node:
            self.left_node.print_tree_info(level + 1)

    def retorna_imovel(self):
        return self.imovel
    
    def calculate_height(self):
        left_height = self.left_node.height if self.left_node else 0
        right_height = self.right_node.height if self.right_node else 0
        return 1 + max(left_height, right_height)

    def update_height(self):
        self.height = self.calculate_height()

    def balance_factor(self):
        left_height = self.left_node.height if self.left_node else 0
        right_height = self.right_node.height if self.right_node else 0
        return left_height - right_height

    def left_rotate(self):
        x = self.right_node
        T2 = x.left_node

        x.left_node = self
        self.right_node = T2

        self.update_height()
        x.update_height()

        return x

    def right_rotate(self):
        y = self.left_node
        T2 = y.right_node

        y.right_node = self
        self.left_node = T2

        self.update_height()
        y.update_height()

        return y

    def balance(self):
        self.update_height()

        balance = self.balance_factor()

        # Left Heavy
        if balance > 1:
            if self.left_node.balance_factor() < 0:
                self.left_node = self.left_node.left_rotate()
            return self.right_rotate()

        # Right Heavy
        if balance < -1:
            if self.right_node.balance_factor() > 0:
                self.right_node = self.right_node.right_rotate()
            return self.left_rotate()

        return self

   
    def print_tree(self, level=0):
        if self.right_node:
            self.right_node.print_tree(level + 1)

        print("  " * level + f"Data: {self.imovel.retorna_data()}, Inquilino: {self.imovel.recebe_inquilino()}")

        if self.left_node:
            self.left_node.print_tree(level + 1)

    def add_imovel(self, new_imovel):
        if new_imovel.retorna_data() == self.imovel.retorna_data():
            return self

        dt_ext = new_imovel.retorna_data()
        dt_ext = datetime.strptime(dt_ext, '%d/%m/%Y')

        dt_curr = self.imovel.retorna_data()
        dt_curr = datetime.strptime(dt_curr, '%d/%m/%Y')

        if dt_ext > dt_curr:
            if self.right_node is None:
                self.right_node = Node(new_imovel)
            else:
                self.right_node = self.right_node.add_imovel(new_imovel)
        else:
            if self.left_node is None:
                self.left_node = Node(new_imovel)
            else:
                self.left_node = self.left_node.add_imovel(new_imovel)

        self = self.balance()

        # informações sobre a árvore que o chat sugeriu, nao acho tao relevante para mostrar
        # print(f"Node with Data {new_imovel.retorna_data()} was balanced. Height: {self.height}, Balance Factor: {self.balance_factor()}")

        return self


# teste search pelo nome
    def search_by_name(self, name):
            results = []
            if self.left_node:
                results.extend(self.left_node.search_by_name(name))

            imovel = self.retorna_imovel()
            if name.lower() in imovel.recebe_inquilino().lower():
                results.append(imovel)

            if self.right_node:
                results.extend(self.right_node.search_by_name(name))

            return results

#teste deletar
    def delete_inquilino_from_json(json_path, inquilino):
        with open(json_path, 'r', encoding='utf-8') as f:
            data = json.load(f)

        for imovel, contratos in data.items():
            contratos_a_remover = [contrato for contrato, detalhes in contratos.items() if detalhes.get("Inquilino", "").lower() == inquilino.lower()]
           
            for contrato in contratos_a_remover:
                del data[imovel][contrato]

        with open(json_path, 'w', encoding='utf-8') as file:
            json.dump(data, file, indent=2)

    # Teste deletar funcionando, mas só funciona dentro da propria funcao
    nome_a_deletar = "RAFAEL R"
    delete_inquilino_from_json('imoveisTeste.json', nome_a_deletar)

    
```

```python
import json

with open('imoveisTeste.json', encoding='utf-8') as f:
    data = json.load(f)

imoveis = []

for imovel_key, imovel_info in data.items():
    imovel_tree = None

    for contrato_key, contrato_info in imovel_info.items():
        data_contrato = contrato_info.get("Data_Contrato_Inicial")
        dias_locados = contrato_info.get("Dias_Locados")
        inquilino = contrato_info.get("Inquilino")
        telefone = contrato_info.get("Telefone")

        novo_imovel = Imovel(
            data_contrato=data_contrato,
            dias_locados=dias_locados,
            inquilino=inquilino,
            telefone=telefone
        )

        if imovel_tree is None:
            imovel_tree = Node(novo_imovel)
        else:
            imovel_tree = imovel_tree.add_imovel(novo_imovel)

    imoveis.append(imovel_tree)
    
    raiz = None
for imovel_tree in imoveis:
    if raiz is None:
        raiz = imovel_tree
    else:
        raiz = raiz.add_imovel(imovel_tree.retorna_imovel())

# para imprimir a arvore de cada imovel
for idx, imovel_tree in enumerate(imoveis):
    print(f"\nÁrvore AVL para Imóvel {idx + 1}:")
    imovel_tree.print_tree_info()
```
