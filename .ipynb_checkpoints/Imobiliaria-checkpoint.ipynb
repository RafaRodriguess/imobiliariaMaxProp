```python
class Imovel:
#Caracteristicas de Imovel: data_contrato, dias_locados, inquilino, telefone
    def __init__(self,data_contrato,dias_locados,inquilino,telefone):

        self.data_contrato = data_contrato
        self.dias_locados = dias_locados
        self.inquilino = inquilino
        self.telefone = telefone

    def retorna_data(self):
        return str(self.data_contrato)

    def recebe_dias_locados(self):
        return str(self.dias_locados)

    def recebe_inquilino(self):
        return str(self.inquilino)

    def recebe_telefone(self):
        return str(self.telefone)

# imovel = Imovel('20/01/2022','120','Joao da Silva','15995624785')
# print(imovel.retorna_data())
# print(imovel.recebe_dias_locados())
# print(imovel.recebe_inquilino())
# print(imovel.recebe_telefone())
```

```python
from datetime import datetime
#Implementando arvore binária:
class Node:

    def __init__(self,imovel):
        self.right_node = None
        self.left_node = None
        self.imovel = imovel

    def retorna_imovel(self):
        return self.imovel
    
    def print_tree(self):
        if self is not None:
            if self.left_node:
                self.left_node.print_tree()

            imovel = self.retorna_imovel()
            print(f"Data: {imovel.retorna_data()}, Inquilino: {imovel.recebe_inquilino()}")

            if self.right_node:
                self.right_node.print_tree()

    def add_imovel(self,new_imovel):
      
        if new_imovel == self.imovel:
            return False

        dt_ext = new_imovel.retorna_data()
        dt_ext = datetime.strptime(dt_ext, '%d/%m/%Y')

        dt_curr = self.imovel.retorna_data()
        dt_curr = datetime.strptime(dt_curr, '%d/%m/%Y')

        if dt_ext > dt_curr:
            if self.right_node is None:
                self.right_node = Node(new_imovel)
            else:
                self.right_node.add_imovel(new_imovel)
        else:
            if self.left_node is None:
                self.left_node = Node(new_imovel)
            else:
                self.left_node.add_imovel (new_imovel) 
```
